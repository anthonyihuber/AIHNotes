```python
from matplotlib import pyplot as mp
import numpy as np

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

x_values = np.linspace(0, 400, 801)
sig = 110#um
mu = 0#um
mp.plot(x_values, gaussian(x_values, mu, sig))

x2 = np.linspace(0, 300, 5)
x3 = gaussian(x2,0,110)

mp.show()
```

```python
mp.plot(x2,np.subtract(1,x3))
```
