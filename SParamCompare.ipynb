```python
dataset1 = '230126_RFCRF1_RT.S2P'
dataset1a = '230126_RFCRF1_RT_OFF.S2P'
dataset2 = '230126_RFCRF2_RT.S2P'
dataset2a = '230126_RFCRF2_RT_OFF.S2P'
dataset3 = '230126_RFCRF3_RT.S2P'
dataset3a = '230126_RFCRF3_RT_OFF.S2P'
```

```python
data1 = rf.Network(CDD + dataset1)
data1a = rf.Network(CDD + dataset1a)
data2 = rf.Network(CDD + dataset2)
data2a = rf.Network(CDD + dataset2a)
data3 = rf.Network(CDD + dataset3)
data3a = rf.Network(CDD + dataset3a)
```

```python
data1.plot_s_db()
```

```python
s11 = data1.s[:,0,0]
s21 = data1.s[:,1,0]
s12 = data1.s[:,0,1]
s22 = data1.s[:,1,1]

p11 = 10**np.log10(np.real(s11*np.conj(s11)))
p12 = 10**np.log10(np.real(s12*np.conj(s12)))
p21 = 10**np.log10(np.real(s21*np.conj(s21)))
p22 = 10**np.log10(np.real(s22*np.conj(s22)))

plt.plot(data1.f,p11,label='S11')
plt.plot(data1.f,p12,label='S12')
plt.plot(data1.f,p21,label='S21')
plt.plot(data1.f,p22,label='S22')

plt.xlim(min(data1.f),max(data1.f))
plt.yscale("log")
plt.legend()
```
