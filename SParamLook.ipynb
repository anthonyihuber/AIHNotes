```python
import skrf as rf
import matplotlib.pyplot as plt
import numpy as np
```

```python
CDD = 'C:/Users/antho/OneDrive/Desktop/School/PhD/PrimeCam/Cooldown S2P Data/'
datasetWarm = '230117_thru_warm.S2P'
datasetCold = '230118_thru_cold.S2P'
```

```python
dataW = rf.Network(CDD + datasetWarm)
dataC = rf.Network(CDD + datasetCold)
```

```python
Warm11 = dataW.s[:,0,0]
Warm21 = dataW.s[:,1,0]
Warm12 = dataW.s[:,0,1]
Warm22 = dataW.s[:,1,1]

Cold11 = dataC.s[:,0,0]
Cold21 = dataC.s[:,1,0]
Cold12 = dataC.s[:,0,1]
Cold22 = dataC.s[:,1,1]
```

```python
P_Warm11 = 10*np.log10(np.real(Warm11*np.conj(Warm11)))
P_Warm21 = 10*np.log10(np.real(Warm21*np.conj(Warm21)))
P_Warm12 = 10*np.log10(np.real(Warm12*np.conj(Warm12)))
P_Warm22 = 10*np.log10(np.real(Warm22*np.conj(Warm22)))

P_Cold11 = 10*np.log10(np.real(Cold11*np.conj(Cold11)))
P_Cold21 = 10*np.log10(np.real(Cold21*np.conj(Cold21)))
P_Cold12 = 10*np.log10(np.real(Cold12*np.conj(Cold12)))
P_Cold22 = 10*np.log10(np.real(Cold22*np.conj(Cold22)))
```

```python
plt.plot(dataW.f/1e9,P_Warm11,label='WARM S11')
plt.plot(dataW.f/1e9,P_Warm12,label='WARM S12')
plt.plot(dataW.f/1e9,P_Warm21,label='WARM S21')
plt.plot(dataW.f/1e9,P_Warm22,label='WARM S22')

plt.plot(dataC.f/1e9,P_Cold11,label='COLD S11')
plt.plot(dataC.f/1e9,P_Cold12,label='COLD S12')
plt.plot(dataC.f/1e9,P_Cold21,label='COLD S21')
plt.plot(dataC.f/1e9,P_Cold22,label='COLD S22')

plt.xlim(min(dataW.f/1e9),max(dataW.f/1e9))
plt.ylabel("Magnitude (dB)")
plt.xlabel("Frequency (GHz)")
plt.title('Thru Measurement')
plt.legend()
```

```python
plt.plot(dataW.f/1e9,P_Warm11,label='S11')
plt.plot(dataW.f/1e9,P_Warm12,label='S12')
plt.plot(dataW.f/1e9,P_Warm21,label='S21')
plt.plot(dataW.f/1e9,P_Warm22,label='S22')

plt.xlim(min(dataW.f/1e9),max(dataW.f/1e9))
plt.ylabel("Magnitude (dB)")
plt.xlabel("Frequency (GHz)")
plt.title('RT Thru Measurement')
plt.legend()
```
